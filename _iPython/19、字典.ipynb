# 数据类型详解-字典

> 字典也是一种数据的集合，由键值对组成的数据集合，字典中的键不能重复
>
> 字典中的键必须是不可变的数据类型，常用的键主要是：字符串，整型。。。

### 字典的定义

+ 字典可以通过将以逗号分隔的 `键: 值` 对列表包含于花括号之内来创建字典
+ 也可以通过 [`dict`](https://docs.python.org/zh-cn/3.7/library/stdtypes.html#dict) 构造器来创建

 `{'jack': 4098, 'sjoerd': 4127}` 或 `{4098: 'jack', 4127: 'sjoerd'}`

```python
# 字典的定义
d1={}
print(type(d1))
d2=dict()
print(type(d2))
d3=dict([['a','1'],['b','2']])
print(d3)


# 字典的新增
d = {'jack': 4098, 'sape': 4139}
d['guido'] = 4127
print(d)
```

```python
# 1。 使用{}定义
vardict = {'a':1,'b':2,'c':2}
print(vardict,type(vardict))

# 2。 使用 dict(key=value,key=value) 函数进行定义
vardict = dict(name='zhangsan',sex='男',age=22)
print(vardict)

# 3。 数据类型的转换  dict(二级容器类型) 列表或元组，并且是二级容器才可以转换
vardict = dict([['a',1],['b',2],['c',3]])  # {'a': 1, 'b': 2, 'c': 3}
print(vardict)

# 4。zip压缩函数，dict转类型
var1 = [1,2,3,4]
var2 = ['a','b','c','d']
print(list(zip(var1,var2)))

# 转换的原理和上面的第三种 是一个原理
var1 = [1,2,3,4]
var2 = ['a','b','c','d']
vardict = dict(zip(var1,var2))  # {1: 'a', 2: 'b', 3: 'c', 4: 'd'}
print(vardict)
```

### 字典的操作

```python
# 因为k：w的形式，字典不可+拼接和*重复
var1 = {'a': 1, 'b': 2, 'c': 3}
var2 = {1: 'a', 2: 'b', 3: 'c', 4: 'd'}
# res = var1 +  var2 # XXXX  TypeError
# res = var1 * 3 # xxxx TypeError

# 获取元素
res = var1['a']
print(res)

# 修改元素
res = var1['a'] = 111
print(res)

# 删除元素
del var1['a']
print(var1)

# 添加元素
var1['aa'] = 'AA'
print(var1)

# 如果字典中的key重复了，会被覆盖
var1['aa'] = 'aa'
print(var1)
```

### 成员检测和获取

```python
var1={'b': 2, 'c': 3, 'aa': 'AA'}
# 三 成员检测和获取  ,只能检测是否存在key，不能检测value
res1 = 'AA' in var1
res2 = 'AA' not in var1
print(res1,res2)

# 获取当前字典的长度 只能检测当前又多少个键值对
res = len(var1)
print(res)

# 获取当前字典中的所有 key 键
res = var1.keys()
print(list(res),res,type(res))

# 获取字典中所有的 value 值
res = var1.values()
print(list(res),res,type(res))

```

### 字典的遍历

```python
# 四， 对字典进行遍历

var1={'a':1,'b': 2, 'c': 3}
# （1）在遍历当前的字典时，只能获取当前的key,但是可以通过对应key的方式取对应的value值
for i in var1:
    print(i) # 只能获取 key
    print(var1[i]) # 通过字典的key获取对应value

print('----'*10)
#（2）遍历字典时，使用 items() 函数，可以在遍历中获取key和value
for k,v in var1.items():
    print(k)  # 遍历时的 key
    print(v)  # 遍历时的 value

print('----'*10)
# (3) 遍历字典的所有key
for k in var1.keys():
    print(k)

print('----'*10)
# (4) 遍历字典的所有 value
for v in var1.values():
    print(v)
```

### 字典的相关函数


- len(字典) 获取字典的键值对个数
- dict.keys()  获取当前字典的所有key 键，组成的列表
- dict.values() 获取当前字典的所有 value 值，组成的列表
- dict.items() 返回由字典项 ((键, 值) 对) 组成的一个新视图
- iter(d) 返回以字典的键为元素的迭代器。


- iter(d) 返回以字典的键为元素的迭代器。 这是 iter(d.keys()) 的快捷方式。
- clear() 移除字典中的所有元素。
- copy() 返回原字典的浅拷贝
- get(key[, default]) 如果 key 存在于字典中则返回 key 的值，否则返回 default。
- reversed(d) 返回一个逆序获取字典键的迭代器。 这是 reversed(d.keys()) 的快捷方式。
- dict.update(),更新字典,如果key存在，则更新，对应的key不存在则添加
- setdefault(key[, default]) 如果字典存在键 key ，返回它的值。如果不存在，插入值为 default 的键 key ，并返回 default 。 default 默认为 None。
- dict.pop(key) 通过 key 从当前字典中弹出键值对  删除
- dict.popitem() LIFO: Last in, First out.后进先出 把最后加入到字典中的键值对删除并返回一个元组

```python
d = {'a':1,'b':2,'c':3}

# iter(d) 返回以字典的键为元素的迭代器。 这是 iter(d.keys()) 的快捷方式。
r=iter(d)
print(list(d))

# clear() 移除字典中的所有元素。
d.clear()
print(d)

d = {
    'name':{'a':1,'b':2,'c':3},
    'pwd':'123456'
}

# copy() 返回原字典的浅拷贝
new_d=d.copy()
print(id(new_d['name']))
print(id(d['name']))

# 字典的深拷贝，同列表
import copy
new_d = copy.deepcopy(d)
print(id(new_d['name']))
print(id(d['name']))

# get(key[, default]) 如果 key 存在于字典中则返回 key 的值，否则返回 default。
res2=d.get('a')
res1=d.get('name')
print(res1,res2)

# reversed(d) 返回一个逆序获取字典键key值的迭代器。 这是 reversed(d.keys()) 的快捷方式。
res= reversed(d)
print(list(res))
```

### 字典推导式

- 和列表推导式类似返回一个字典 d={x: x**2 for x in range(10)}
- 注意和集合推导式区分{x**2 for x in range(10)}

```python

# 把字典中的键值对位置进行交换 {'a':1,'b':2,'c':3}
vardict = {'a':1,'b':2,'c':3}

# 普通方法实现 字典中的键值交换  {1: 'a', 2: 'b', 3: 'c'}
newdict = {}
for k,v in vardict.items():
    newdict[v] = k
# print(newdict)

# 使用字典推导式完成  {1: 'a', 2: 'b', 3: 'c'}
newdict = {v:k for k,v in vardict.items()}
# print(newdict)

# 注意：以下推导式，返回的结果是一个集合，集合推导式
# newdict = {v for k,v in vardict.items()}
# print(newdict,type(newdict))


# 把以下字典中的是偶数的值，保留下来，并且交换键值对的位置
vardict = {'a':1,'b':2,'c':3,'d':4}

# 普通方式完成  {2: 'b', 4: 'd'}
# newdict = {}
# for k,v in vardict.items():
#     if v % 2 == 0:
#         newdict[v] = k
# print(newdict)

# 字典推导式完成  {2: 'b', 4: 'd'}
newdict = {v:k for k,v in vardict.items() if v % 2 == 0}
# print(newdict)
```
