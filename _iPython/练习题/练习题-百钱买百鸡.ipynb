```python
# 思路：高中方程思想？
# x、y、z代表三种鸡的个数，xyz大于零且有上限，然后x+y+z =100，总花费 cost =100

count = 0
num = 0
for x in range(100//3):
    for y in range(100//1):
        for z in range(int(100//0.5)):
            num +=1
            if x+y+z ==100 and x*3+y+z*0.5 ==100:
                count +=1
                print(f'供给{x}只，母鸡{y}只，小鸡{z}只。')
                
print(f'\n共有{count}种方案')
print(f'\n共计算{num}次')
```

```python
# 简单优化：可以直接去掉最多次数的那层循环，因为确定了xy，那么z的值也就确定了

count = 0
num = 0
for x in range(100//3):
    for y in range(100//1):
        z=100-x-y
        num +=1
        if x+y+z ==100 and x*3+y+z*0.5 ==100:
            count +=1
            print(f'供给{x}只，母鸡{y}只，小鸡{z}只。')
                
print(f'\n共有{count}种方案')
print(f'\n共计算{num}次')
```
