```python
# 
# xyzxyzx+y+z =100 cost =100

count = 0
num = 0
for x in range(100//3):
    for y in range(100//1):
        for z in range(int(100//0.5)):
            num +=1
            if x+y+z ==100 and x*3+y+z*0.5 ==100:
                count +=1
                print(f'{x}{y}{z}')
                
print(f'\n{count}')
print(f'\n{num}')
```

```python
# xyz

count = 0
num = 0
for x in range(100//3):
    for y in range(100//1):
        z=100-x-y
        num +=1
        if x+y+z ==100 and x*3+y+z*0.5 ==100:
            count +=1
            print(f'{x}{y}{z}')
                
print(f'\n{count}')
print(f'\n{num}')
```
