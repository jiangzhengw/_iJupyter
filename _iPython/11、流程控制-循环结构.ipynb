### 循环结构

```python
# while 循环

num =0

# while死循环
# while True:
#     print('number：',num)
# 有条件循环
while num <=10:
    print('number：',num)
    num +=1
print("Final num ：",num)
```

```python
# for in 循环 一般用于便利容器类型的数据

s = '12345'
for i in s:
    print(i)
    
print(type(range(1,10)))

for i in range(1,5):
    print(i)
```

1、break 结束跳出,不会执行下一次循环

2、continue 跳过本次循环，会执行下一次循环

3、pass 占位

quit()
exit()是用来终止python程序的
    

```python
num=0
while num <10:
    num+=1
    if num%2 ==0:
#         print('%d为偶数,跳过本次循环'%num)
        continue
    else:
        print('%d为基数'%num,)
    
    if num == 5:
        break
        
```

```python
# 思路：定义两个不同颜色的初始星号，然后使用while循环嵌套两次实现打印10*10的矩形，判断偶数和基数的时候打印颜色不同的星号。
# print('☆', end='')
# print('★',end = '')

i = 0
while i < 10:
    j =0
    while j<10:
        print('★',end = '')
        j+=1
    print('\n', end='')
    i+=1
    
print("--------------")
for i in range(10):
    for j in range(10):
        print('★',end = '')
    print("")
```

```python
# 隔列打印
i = 0
while i < 10:
    j =0
    while j<10:
        if j%2 ==0:
            print('★',end = '')
        else:
            print('☆', end='')
        j+=1
    print('\n', end='')
    i+=1

print("---------------------")

for i in range(10):
    for j in range(10):
        if j%2 ==0:
            print('★',end = '')
        else:
            print('☆', end='')
    print('\n', end='')
    
print("-----------------------")


i = 0
while i < 100:
    if i%2 == 0:
        print('★',end = '')
    else:
        print('☆', end='')
    if i%10 == 9:
        print('\n',end = '')
    i += 1
```

### 打印三角形或者菱形

思路：打印二维图形使用两个参数i、j控制行数和列数，然后内部详细控制打印的规则，0-i ，0-（10-i）等，具体详细思考

```python
# 打印三角形

# 控制行数
row =10

# 左上右下等腰直角
for i in range(row):
    for j in range(0,row-i):
        print('☆',end = '')
    for j in range(row-i,row):
        print('★',end = '')
    print("")
 
print("-------分界线-------1")

for i in range(row):
    for j in range(0,row-i):
        #两个空格
        print('  ',end = '')
    for j in range(row-i,row):
        print('★',end = '')
    print("")
 
print("-------分界线-------2")

for i in range(row):
    for j in range(0,row-i):
        print('☆',end = '')
#     for j in range(row-i,row):
#         print(' ',end = '')
    print("")
 
print("-------分界线-------3")

for i in range(row):
    for j in range(0,row-i-1):
        print(' ',end = '')
    for j in range(row-i-1,row):
        print('★',end = '')
    print("")
 
print("-------分界线-------4")

# 坐下右上等腰直角
for i in range(row):
    for j in range(0,i):
        print('☆', end='')
    for j in range(i,row):
        print('★',end = '')
    print("")

print("-------分界线-------5")

for i in range(row):
    for j in range(0,i):
        #两个空格
        print('  ', end='')
    for j in range(i,row):
        print('★',end = '')
    print("")

print("-------分界线-------6")

for i in range(row):
    for j in range(0,i):
        print('☆', end='')
#     for j in range(i,row):
#         print(' ',end = '')
    print("")

print("-------分界线-------7")

for i in range(row):
    for j in range(0,i):
        #一个空格
        print(' ', end='')
    for j in range(i,row):
        print('★',end = '')
    print("")
```

```python
# 打印菱形

# 第一个思路两个上下等腰就是竖着的菱形
raw = 10

for i in range(raw):
    for j in range(0,raw-i-1):
        print(' ',end='')
    for j in range(raw-i-1,10):
        print('★',end = '')
    print("")
for i in range(raw):
    for j in range(0,i):
        print(' ',end='')
    for j in range(i,10):
        print('★',end = '')
    print("")

# 横着的菱形，思路相同，先考虑实现横着的等腰三角形

# TODO
```
