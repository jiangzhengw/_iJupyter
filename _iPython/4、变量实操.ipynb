### 整数、浮点的多则运算


```python
a,b = 15,7
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

### 字符串的拼接

只有字符串和字符串可以拼接

```python
s1 = 'hello'
s2 = 'python'
s3 = '!'
a1 = '姜正炜'
a2 = 24
a3 ='岁'

print(s1+s2+s3)
print(a1+str(a2)+a3)
print('{}{}{}'.format(s1,s2,s3))
print('{}{}{}'.format(a1,a2,a3))
```

### 占位符

```python
name,age = '黎晟',24

# 方法1:适合简单且有序的占位
print("姓名：%s,年龄：%d"%(name,age))

# 方法2：适合一个量多次占位
print("姓名：{0},年龄：{1}".format(name,age))

# 方法3：适合有明确的映射关系的占位
print(f"姓名：{name},年龄：{age}")

```
