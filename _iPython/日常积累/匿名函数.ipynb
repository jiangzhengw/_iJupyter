

- 匿名函数本身是函数需要，赋值后需要函数调用
- 匿名函数不需要return来返回值，表达式本身结果就是返回值。
+ lambda语句中，冒号前是参数，可以有多个，用逗号隔开，冒号右边的返回值

```python
import platform as plt
plt.python_version()


from icecream import ic

# 无参匿名函数:
t = lambda : True 
print(t())
```

```python
# 有参数匿名函数
f = lambda a,b,c:a+b+c  
print(f(1,2,3))
```

```python
# 字符串形参匿名函数
s= lambda s:''.join(s.split())
print(s("this is\na test"))
```

```python
# 带收集参数：接收元组
m = lambda *args:args
print(m(2,3,4))
```

```python
# 带默认参数：关键字参数
m = lambda a,b=1,c=2:a*b*c
print(m(2))
print(m(2,3,4))
```

```python
# 带关键字收集参数,接收字典
n = lambda **kwargs: kwargs
print(n(x=1,y=2,z=3))
```

```python
#直接后面传递实参

l =(lambda x,y:x if x>y else y) (1,2) 
i = (lambda s:''.join(s.split()))('this is\na test')

# 不需要传参，因为l和i代表的已经是函数的返回值。
print(l)
print(i)
```

```python
#lambda返回的值，结合map,filter,reduce使用
from functools import reduce

r = filter(lambda x:x%2==1,[1,2,3,4,5,6,7])
r2 = list(map(lambda x,y:x+y,[1,2,3,4],[4,3,2,1]))
r3 = reduce(lambda x,y:x+y,[1,2,3,4,5])

# 等价于下面的列表推导式
# r = [x for x in [1,2,3,4,5,6] if x%3==0]
print(type(r))

# 打印一个迭代器 iterator
# sorted() 函数对所有可迭代的对象进行排序操作，返回排序后的列表。
# print(sorted(r))

print(list(r))
# print(tuple(r))
print(r2)
print(r3)

# for i in r:
#     print(i,end=' ')
```

```python
# 嵌套使用

#lambda嵌套到普通函数中,lambda函数本身做为return的值

def increment(n):
     return lambda x: x+n
    
# print(type(increment(2)))
f = increment(4)
f(2)

def say():
    title = 'Sir,'
    action= lambda x: title + x
    return action
g =say()
print(g('i am tired!'))
```
