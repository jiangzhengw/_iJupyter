### 算数运算符

加 减 乘 除 商 余 幂

```python
a,b = 3,2
print(a+b) #加
print(a-b) #减
print(a*b) #乘
print(a/b) #除
print(a//b) #商
print(a%b) #余
print(a**b) #幂
```

### 字符串运算

+ 字符串与数字不能直接参与运算
+ 字符串+字符串 记过是字符串的拼接
+ 

```python
s1 = 'hello'
s2 = 'python'
s3 = '!'
a1 = '姜正炜'
a2 = 24
a3 ='岁'

print(s1+s2+s3)
print(a1+str(a2)+a3)
print('{}{}{}'.format(s1,s2,s3))
print('{}{}{}'.format(a1,a2,a3))
print(f'{s1} {s2} {s3}{s3}{s3}')
```

### 赋值运算符

```python
# 运算顺序为从右到左
i=2+4
print(i)

# 支持链式赋值 
a=b=c=2000
print(a,id(a))
print(b,id(b))
print(c,id(c))

# 支持参数赋值 += -+ *= /= //= %= **=
a=2
a+=5
print("a:",a)
a-=5
print("a:",a)
b=2
b-=5
print("b:",b)
c=2
c*=5
print("c:",c)
d=2
d/=5
print("d:",d)
e=11
e//=5
print("e:",e)
f=12
f%=5
print("f:",f)
g=2
g**=3
print("g:",g)

# 支持系列解包赋值 
a,b,c=20,30,40
print(a,b,c)
print(id(a),id(b),id(c))
a,b,c=c,a,b
print(a,b,c)
print(id(a),id(b),id(c))
```

### 比较运算符

对变量或表达式的结果进行大小、真假比较

```python
# 比较运算符：
# >,<,>=,<=,!=
a=2000
b=2000
print(a>b,a<b,a>=b,a<=b,a!=b)

# == ：对象value的比较
print(a==b)

# is,is not ：对象id的比较（内存地址的比较）
print(id(a),id(b))
print(a is b,a is not b)

print("\n---->分界线<----\n")
a,b=2519,2519
print(a>b,a<b,a>=b,a<=b,a!=b)

# == ：对象value的比较
print(a==b)

# is,is not ：对象id的比较（内存地址的比较）
print(id(a),id(b))
print(a is b,a is not b)

```

### 布尔运算符，逻辑运算符


```python
# and,or,not,in,not in

a,b=1,2
print(a==1 and b==2)
print(a==1 and b!=2)
print(a==1 or b==3)
print(not a==1)
print(a in [a,b])
print('h' not in 'hello python')
print(40 and 0)
print(40 and 30)
print(40 or 30)
```

### 位运算符

将数据转成二进制然后进行计算

```python
print(1&2)
print(1&9)
print(6|4)
print(4<<2) # 二进制数左移两位等于乘以2的2次方
print(4>>2) # 二进制数右移两位等于除以2的2次方
```

#### 按位异或运算 ^ 
当对应二进位想异时结果为1

```python
print(0b1011 ^ 0b1100) # == 0b111 ==4+2+1 ==7
```

#### 按位取反运算 ~ 
对应位数二进制取反运算
### ~~是取反两次，作用是去掉小数部分，因为二进制都是整数：
除了~~n 还可以用
n<<0
n>>0
n|0



```python
print(~0b11000) # == -(24+1)
```
