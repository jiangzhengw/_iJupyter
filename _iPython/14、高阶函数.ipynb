### 递归函数

```python
# 递归函数就是定义了一个函数，然后在函数内，自己调用了自己这个函数

def do(num):
    print(num)
    if num >0:
        do(num-1)
        
    print(num)
    
    
do(3)
```

### 回调函数


- 定义一个函数，函数中的一个参数要求是另一个函数
- 带回调函数参数的函数

```python
# 回调函数为普通函数
def func(f):
    f()
    
def love():
    print('I love you!')
    

func(love)
```

```python
# 回调函数为内置函数
def func2(x,y,z,f):
    print(f([x,y,z]))

func2(1,2,3,sum)
```

### 闭包



```python
# 全局变量会出现的问题，如果money突然清零，那所有的操作又需要重新开始
money = 0

def fun1():
    global money
    money +=100

def fun2():
    global money
    money +=200
    
def fun3():
    global money
    money -=50
    
fun1()
fun1()
fun1()
fun2()
fun3()
print(money)
```

```python
# 将以上代码进行闭包改造

def person():
    money=0
    def work(h):
        nonlocal money # 在内函数使用了外函数中的局部变量
        money +=100*h
        print(money)
    def over_time(h):
        nonlocal money
        money +=200*h
        print(money)
    def coast(times):
        nonlocal money
        money -=50*times
        print(money)
    return work #在外函数反悔了一个内函数，work就是闭包函数

target = person()
target(10)
print(person.__closure__)
print(target.__closure__)
```
