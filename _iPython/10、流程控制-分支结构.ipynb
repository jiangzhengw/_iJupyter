```python
# 思路：定义一个初始年份，然后计算输入年份和初始年份的差值，然后对12求余，对应的余数输出对应的生肖。

year = int(input('请输入一个四位数年份：'))
if year%12 == 0:
    print("%s年是猴年！"%year)
elif year%12 == 1:
    print("%s年是鸡年！"%year)
elif year%12 == 2:
    print("%s年是狗年！"%year)
elif year%12 == 3:
    print("%s年是猪年！"%year)
elif year%12 == 4:
    print("%s年是鼠年！"%year)
elif year%12 == 5:
    print("%s年是牛年！"%year)
elif year%12 == 6:
    print("%s年是虎年！"%year)
elif year%12 == 7:
    print("%s年是兔年！"%year)
elif year%12 == 8:
    print("%s年是龙年！"%year)
elif year%12 == 9:
    print("%s年是蛇年！"%year)
elif year%12 == 10:
    print("%s年是马年！"%year)
elif year%12 == 11:
    print("%s年是羊年！"%year)
else:
    print("请输入正确年份！")
```
