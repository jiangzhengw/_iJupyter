### 变量的多对多赋值

```python
a,b = '1',2
print(a,b)
print(type(a),type(b))
```

```python
# 用法，使用场景
def fun():
    return "abc",123
a,b=fun()
print(a,b)
```

```python
# python会认为简单的字符串为一样，会指向同一内存地址。节省了内存空间
# 仅仅针对简单的字符串  包含 -- 中文 空格 的不会共享内存

a='aaa'
b='aaa'
o='aaa-'
p='aaa-'
print(id(a),id(b))
print(id(o),id(p))

# 整数的共享 -5 到 256
c= -5
d= -5
e=-6
f=-6
print(id(c),id(d))
print(id(e),id(f))
```

### 变量的多标签

```python
# 只是添加了标签，两个标签指向了同一内存地址
bed='aaa'
bed2=a
print(bed,bed2)
print(id(bed),id(bed2))
```

```python
# 如果重新赋值，会重新打开一个箱子，不会影响另一个标签和内存地址的映射关系

bed='bbb'
print(bed,bed2)
print(id(bed),id(bed2))
```

### 变量的赋值，活用tips

```python
# 复用标签名,会重新新建一个箱子（映射关系），python会自动检测内存空间的映射关系，定时删除无用的内存空间

name='ls'
print(id(name))
name='ls2'
print(id(name))
```
