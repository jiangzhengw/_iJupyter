```python
print(chr(0b100111001011000))
s='乘'
```

```python
# 0b、0o、0x 标识二进制八进制十六进制
# bin() oct() hex()
s='乘' 
print(int(0b1000)) # int()将其他进制转换为十进制
print(int(0o20)) # int()将其他进制转换为十进制
print(int(0x20)) # int()将其他进制转换为十进制
print(ord(s),type(ord(s))) # ord()将字符串转为十进制
# print(int(ord(s),16))
print(bin(ord(s))) # bin()转换为二进制
print(oct(ord(s))) # oct()转换为八进制
print(hex(ord(s))) # hex()转换为十六进制
print(0x10) # == print(int("10",16))
print(int("10",16)) # 以十六进制打印
```
